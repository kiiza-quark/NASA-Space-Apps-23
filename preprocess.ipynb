```python
import pandas as pd
import numpy as np
import os
```

```python

folder_path = "./data/dsc_fc/"

downsampled_dfs = []

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        
        data = pd.read_csv(file_path, delimiter=',', parse_dates=[0], infer_datetime_format=True, na_values='0', header=None)
        data.set_index([0], inplace=True)
        data_3h = data.resample('1.5H').mean()
        data_3h_downsampled = data_3h[::2]
        
        downsampled_dfs.append(data_3h_downsampled)

merged_df = pd.concat(downsampled_dfs)

merged_df.to_csv('./data/dsc_fc/merged_data.csv')
```

```python
kp_df = pd.read_csv('kp_data.csv')
kp_df.head()
```

```python
kp_df['Datetime'] = kp_df.apply(lambda row: f"{int(row['Year'])}-{int(row['Month']):02d}-{int(row['Day']):02d} {int(row['Hour_start']):02d}:00:00", axis=1)

kp_df['Datetime'] = pd.to_datetime(kp_df['Datetime'], format='%Y-%m-%d %H:%M:%S')
```

```python
kp_df = kp_df[['Datetime', 'kp_index_x3']]
kp_df
```

```python
kp_df.set_index('Datetime', inplace=True)
```

```python
merged_kp_fc = pd.merge(merged_df, kp_df, left_index=True, right_index=True)
merged_kp_fc
```

```python
merged_kp_fc.to_csv('fc_kp.csv')
```
